```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/all_data.csv')
```

```python
# 수치형 열들만 선택하고 분산 계산
numeric_cols = df.select_dtypes(include=[np.number])
variances = numeric_cols.var()

# 분산이 가장 큰 상위 5개 열 선택
top_5_vars = variances.nlargest(5)

# 히스토그램 그리기
plt.figure(figsize=(15, 10))

for i, col in enumerate(top_5_vars.index):
    plt.subplot(2, 3, i + 1)
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'{col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
